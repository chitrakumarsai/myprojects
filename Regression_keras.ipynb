```python
import pandas as pd
import numpy as np
```

```python
concrete_data = pd.read_csv('FuelConsumption.csv')
concrete_data.head()
```

```python
concrete_data.shape
```

```python
concrete_data = concrete_data.describe()
```

```python
concrete_data.isnull().sum()
```

```python
concrete_data_columns = concrete_data.columns
predictors = concrete_data[concrete_data_columns[concrete_data_columns != 'CO2EMISSIONS']]
target = concrete_data['CO2EMISSIONS']
```

```python
predictors.head()
```

```python
target.head()
```

```python
predictors_norm = (predictors - predictors.mean())/predictors.std()
predictors_norm.head()
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
n_cols = predictors_norm.shape[1]
```

```python
def regression_model():
    model = Sequential()
    model.add(Dense(50, activation = 'relu', input_shape =(n_cols,) ))
    model.add(Dense(50,activation = 'relu'))
    model.add(Dense(1))

    model.compile(optimizer = 'adam', loss = 'mean_squared_error')
    return model
```

```python
model = regression_model()
model.fit(predictors_norm, target, validation_split = 0.2, epochs =100, verbose = 2)
```

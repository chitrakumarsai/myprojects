```python
import numpy as np
import pandas as pd
from math import sqrt
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
movies_df = pd.read_csv('movies.csv')
rating_df = pd.read_csv('ratings.csv')
movies_df.head()
```

```python
movies_df['year'] = movies_df.title.str.extract('(\(\d\d\d\d\))',expand=False)
movies_df['year'] = movies_df.year.str.extract('(\d\d\d\d)',expand=False)
movies_df['title'] = movies_df.title.str.replace('(\(\d\d\d\d\))', '')
movies_df['title'] = movies_df['title'].apply(lambda x: x.strip())
movies_df.head()
```

```python
movies_df['genres'] = movies_df.genres.str.split('|')
movies_df.head()
```

```python
movies_withgenres_df = movies_df.copy()
for index, row in movies_df.iterrows():
    for genre in row['genres']:
        movies_withgenres_df.at[index, genre] = 1
movies_withgenres_df = movies_withgenres_df.fillna(0)
movies_withgenres_df.head()
```

```python
rating_df = rating_df.drop('timestamp', 1)
rating_df.head()
```

```python
userInput = [
            {'title': 'Breakfast Club, The', 'rating': 5},
            {'title': 'Toy Story', 'rating': 3.5},
            {'title': 'Jumanji', 'rating': 2},
            {'title': 'Pulp Fiction', 'rating': 5},
            {'title': 'Akira', 'rating': 4.5}
        ]
inputMovies = pd.DataFrame(userInput)
inputMovies
```

```python
inputID = movies_df[movies_df['title'].isin(inputMovies['title'].tolist())]
inputID
```

```python
inputMovies = pd.merge(inputID, inputMovies)
inputMovies
```

```python
inputMovies = inputMovies.drop('genres',1).drop('year',1)
inputMovies
```

```python
userMovies = movies_withgenres_df[movies_withgenres_df['movieId'].isin(inputMovies['movieId'].tolist())]
userMovies
```

```python
userMovies = userMovies.reset_index(drop =True)
userMovies
```

```python
inputMovies['rating']
```

```python
userGenre_table = userMovies.drop('movieId',1).drop('title',1).drop('year',1).drop('genres',1)
userGenre_table
```

```python
user_profile = userGenre_table.transpose().dot(inputMovies['rating'])
user_profile
```

```python
genreTable = movies_withgenres_df.set_index(movies_withgenres_df['movieId'])
genreTable
```

```python
genreTable = genreTable.drop('year',1)
genreTable.head()
```
